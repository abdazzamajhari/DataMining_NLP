## Cleaning, CaseFolding, Tokenizing, Filtering, dan Stemming

```python
import re, string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

!pip install Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
def cleaningText(text):
    text = re.sub(r"b'", '', text) #menghapus b'
    text = text.replace('\n', '') #menghapus baris baru
    text = re.sub(r'@[A-Za-z0â€“9]+', '', text) # Menghapus @mentions
    text = re.sub(r'#[A-Za-z0â€“9]+', '', text) # Menghapus '#' hash tag
    text = re.sub(r'RT[\s]+', '', text) # Menghapus RT
    text = re.sub(r"http\S+", '', text) # Menghapus link
    text = re.sub(r'[0-9]+', '', text) # Menghapus angka
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.strip(' ')
    return text
```

```python
def casefoldingText(text):
    text = text.lower()
    return text
```

```python
def tokenizingText(text):
    text = word_tokenize(text)
    return text
```

```python
def filteringText(text):
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text
```

```python
def stemmingText(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text
```

## Sentiment Polarity

```python
# sentiment polarity digunakan untuk menentukan orientasi sentimen yang diungkapkan dari data Tweet
# menentukan apakah teks mengungkapkan suatu perasaan yaitu positif, negatif, atau netral dari pengguna yang melakukan postingan twitter

lexicon_positive = dict()

import csv
with open('lexicon_positive.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    for row in reader:
        lexicon_positive[row[0]] = int(row[1])
        
#########
        
lexicon_negative = dict()

import csv
with open('lexicon_negative.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    for row in reader:
        lexicon_positive[row[0]] = int(row[1])
        
#########

def sentiment_analysis_lexicon_indonesia(text):
    score = 0
    for word in text:
        if(word in lexicon_positive):
            score = score + lexicon_positive[word]
    for word in text:
        if(word in lexicon_negative):
            score = score + lexicon_negative[word]
            
    polarity = ''
    if(score > 0):
        polarity = 'positive'
    elif (score < 0):
        polarity = 'negative'
    else:
        polarity = 'neutral'
    return score, polarity
    
    
```

```python
results = tweets['text_preprocessed'].apply(sentiment_analysis_lexicon_indonesia)
results = list(zip(*results))

tweets['polarity_score'] = results[0]
tweets['polarity'] = results[1]
print(tweets['polarity'].value_counts())

tweets
```

# 4 Modeling

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

pipeline_ls = make_pipeline(CountVectorizer(tokenizer = RegexpTokenizer(r'[A-Za-z]+').tokenize,
                                           lowercase = False), SVC(probability=True))
pipeline_ls
```

```python
# pembagian dataset ke train dan test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tweets_new.text_clean, tweets_new.polarity, test_size = 0.3)
```

# 5 - Evaluation 

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import seaborn as sns

print('Training Accuracy :', pipeline_ls.score(X_train, y_train))
print('Testing Accuracy :', pipeline_ls.score(X_test, y_test))
con_mat = pd.DataFrame(confusion_matrix(pipeline_ls.predict(X_test), y_test),
                       columns = ['Predicted:Bad', 'Predicted:Good'],
                       index = ['Actual:Bad', 'Actual:Good'])
                       
print('\nCLASSIFICATION REPORT\n')
print(classification_report(pipeline_ls.predict(X_test), y_test,
                            target_names =['Bad','Good']))
      
print('\nCONFUSION MATRIX')
plt.figure(figsize= (6,4))
sns.heatmap(con_mat, annot = True,fmt='d',cmap="YlGnBu")
                       
```

```python
from matplotlib import pyplot
from sklearn.metrics import roc_auc_score, roc_curve

# generate a no skill prediction (majority class)
ns_probs = [0 for _ in range(len(y_test))]

y_test2 = y_test.map({'positive': 1, 'negative': 0}).astype(int)

# predict probabilities
svm_probs = pipeline_ls.predict_proba(X_test)

# keep probabilities for the positive outcome only
svm_probs = svm_probs[:, 1]

# calculate scores
ns_auc = roc_auc_score(y_test2, ns_probs)
svm_auc = roc_auc_score(y_test2, svm_probs)

# summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('SVM: ROC AUC=%.3f' % (svm_auc))

# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_test2, ns_probs)
svm_fpr, svm_tpr, _ = roc_curve(y_test2, svm_probs)

# plot the roc curve for the model
pyplot.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
pyplot.plot(svm_fpr, svm_tpr, marker='.', label='SVM')

# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')

# show the legend
pyplot.legend()

# show the plot
pyplot.show()
```
